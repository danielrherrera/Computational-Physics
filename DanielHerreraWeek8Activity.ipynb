### Assaignment 2

```python
x = np.arange(20)
```

```python
y = x #Linear
z = x **2  #Quadratic
v = x**3 #Cubic
m = np.exp(x) #Exponential

plt.plot(x, y, 'g', label='Linear')
plt.plot(x, z, 'b', label = 'Quadratic')
plt.plot(x, v, 'y', label = 'Cubic')
plt.plot(x, m, 'm', label = 'Exponential')

plt.xlabel('Values to 20')
plt.ylabel('Corresponding values of x')
plt.legend(loc = 'upper left')
```
